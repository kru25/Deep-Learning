## Neural Networks: A Brief History

````python
Neuron is the basic fundamental unit of nervous system present in brain responsible for all the 
human actions and the responses.Scientists have always wondered what would the structure of a neuron 
be like that it is able to pass and process such complex information within a small time interval.
In 1943 , Scientists Warren McCulloch and Walter-Pitts thus came with a mathematical model for an 
artificial neuron.It follows the concept of all or none characteristic of human nervous system.The
first artificial neuron was called "Perceptron".It was found that single layer of Perceptron cannot be
trained for simple function such as XOR but multiple layers can.Thus neural networks came into picture
.Thus came the concept of Parellel Distributed Programming(PDP).Later using multiple layers of neurons
though could solve many problems but were also considerably slow.With the advancements this problem is
also solved and today we have effective computer programs,hardwares which can train multiple layers of
neurons very fast and produce accurate results.Old Scientists have describe Neural Networks as 
something which is capable of identifying complex problems,processing them without any human efforts.
````

### Running Your First Notebook

````python
Interesting yet funny that we do not know anything about Deep Learning yet and we are going to run our
first deep learning model for Image classification.As mentioned before we will be building an Image 
classifier to identify cats and dogs from a Cats vs Dogs Dataset.There are many such datasets which 
are available online which can be downloaded easily for our model building.
````

````python
Now,the code is already provided to us and we are just required to run the next code cell.Let us try
to comprehend the meaning of the code.
Fastai is divided into four different modules according to the deep learning applications in vast
domains including computer vision,language processing,tabular data and collaborative filtering.
Here since we are building an image classifier we import all classes and objects from the computer 
vision module of fastai library.
Next we need our dataset for cats and dogs.So we download it using the untar_data and decompress the 
images.untar_data downloads the dataset and returns the path of the decompressed images.
Next we need is the labels for our images which in this case is provided by the first letter of the 
filename which is capital in case its a cat else it is a dog.So we define a label function which 
returns a boolean output for if the image is that of a cat or not.
Now we define a DataLoader which is a very efficient way of storing training data in fastai.It gets 
the image data using path returned earlier,splits data into validation set(valid_pct) and random 
seed is also set for reproduciblity,also transforms the images into same size using Resize property
of item_tfms attribute.
Next we use a pretrained model(resnet34) which is already trained on millions of images to train our 
learner and a metric error_rate to evaluate the model performance.This serves as a baseline model.
The pre trained model is then fine tuned by providing parameters such as no of epochs and learning 
rate to improve our predictions.We would learn more about these terms in next lessons.Let us consider
these as the basic steps for training any deep learning model.
1.import necessary libraries or modules.
2.Download and decompress the data and get its path.
3.Get the images using DataLoaders(validation-split,get labels,image transformation)
4.create learner using pretrained neural network architecture
5.fine tune the model for some epochs to improve the model accuracy or error_rate.
````

````python
# CLICK ME 
#First program on the classification of pets images into cats and dogs from pets dataset
from fastai.vision.all import * #importing everything from the fastai.vision module
path = untar_data(URLs.PETS)/'images'#Downloads the dataset pets and decompresses it.
#Inbuilt access to many datasets by fast.ai
#Returns the path of the decompressed images

def is_cat(x): return x[0].isupper()#Label function defined by if the first letter of the filename is 
#capital(cat) or not(else dog)

dls = ImageDataLoaders.from_name_func( #see doc for this command below to see which hyperaparameters to be passed
    path, get_image_files(path), valid_pct=0.2, seed=42,
    label_func=is_cat, item_tfms=Resize(224),num_workers=0)
#CNN(convolutional neural network)
#learner for training the data.Pass the dataset,neural architecture(resnet34 for computer vision) and error rate
learn = cnn_learner(dls, resnet34, metrics=error_rate) #error rate is calculated using valid_pct to prevent overfitting
learn.fine_tune(1) #finetuning the learner by running several epochs
````

### What Is a Neural Network?

````python
Handling categorical or numeric data as input seems easy as we can define some relationship or some 
equation to relate our input and output data using some equation.But when it comes to images how do we
define the relationship or extract features from the images.Arthur Samuels,an IBM researcher had 
mentioned in his research that if we program computers to extract the information from examples and
then make predictions using that information.This was his idea of Machine Learning then which we now 
call training our models.This idea only gave birth to deep learning and in one of his works he quoted 
that if instead of making predictions once we can provide information to the computer in such a way 
that everytime it will evaluate the model performance on test data and accordingly change some
parameters which he called weights and keeps repeating the process until we get best performance of 
the model.Samuel had quoted that this process can be made automatic and that computer learns from its
performance.For this we would need some mechanism to update the weights.One mechanism could be to take
the difference between the weights of previous models which would give us a direction to update our
weights in subsequent later models.This is the main concept of Deep learning and A neural network 
model provides us a way to automate the process of weight updation and thus building models.
````

````python
As discussed earlier neural networks are mathematical models for brains neurons.They are very flexible
such that they can solve very complex problems in small time.The process of training neural networks 
involves a method for weights updation.These methods or alogorithms are called optimizers and one of 
the most common one used is Stochastic Gradient Descent(SGD) about which we will discuss in detailed 
in next lessons and we would also learn about building a SGD from scratch.
But for now its fully automatic and is handled by our programs.
````

````python
So coming back to our input and output data we have images as our input, we feed them into our neural
net model which gives us predictions in the form of cats or dogs.Seems like machine learning no.It is
Machine Learning actually where we are evaluating our model performance on a test data using metric 
such as accuracy.Some of the terminology used earlier and how its different from the ones we use today
are as follows:-
1.Model is called architecture.
2.Weights are termed as parameters.
3.Results are called predictions.
4.Measure of performance is called loss.
5.Loss depends on  targets as well as predicted variable.
````

### What Our Image Recognizer Learned

````python
If we talk about the structure of our neural nets,it consists of several layers of neurons interconn-
ected to each other using networks so that information is passed forward.But is the same information
passed through every layer then how do we arrive at our prediction till the last layer?
So every layer actually contributes to our prediction and helps in identifying a unique feature about
the image.So initial layer helps in identifying the background,texture,horizontal vertical edges in 
the image.Next subsequent layer helps in finding 2D shapes such as circles,corners,squares,lines etc.
Further layer again helps in objects such as cars,wheels,flowers,text etc.Using this information 
further layers can identify higher concepts and at the end we have the final prediction.So information
passed from every layer is different.
Some models even have 100s,1000s of layers in them if the images or our data consists of many objects.
Fastai generally practises Transfer Learning that is using a pretrained model to train and finetune 
our models.For cats vs dogs also we used resnet34 which is a pretrained model already trained on Image
Dataset containing millions of images.
````

## Deep Learning Is Not Just for Image Classification

````python
Till now we discussed about Image classification but does Deep learning have only this application.It
is also used for tasks like identifying malignant tumors.
Everyone knows about the innovation of self driving cars by google.Not many know that it uses Machine 
Learning and Deep learning for automation.How the car would know where it has to slow down or where it
is supposed to stop when there is someone on road?It does so by identifying objects in the picture 
which is done using deep learning.We do Image Segmentation to analyze every pixel in the image.Here we
build a model using fastai using a dataset from one of the papers on Semantic Object Classes in Video.
````

````python
We can observe that here also we have followed the same steps of untar_data,Loading the data in 
DataLoaders,creating a learner using pretrained architecture and then using learner to fine_tune
the model till 8 epochs.We get the path using untar_data.Then we load the data in Dataloaders and one 
of the differences here is that data is stored in batch size of 8 images.We would learn more about 
batch sizes in further lessons for now it increases the reproduciblity of our model while optimizing
weights.For image classification we used cnn_learner,here we are using unet_learner which is mainly 
used for semantic segmentation.We
````

````python
#Image Segmentation using fastai 
path = untar_data(URLs.CAMVID_TINY)#Downloading and decompressing dataset
dls = SegmentationDataLoaders.from_label_func(
    path, bs=8, fnames = get_image_files(path/"images"),
    label_func = lambda o: path/'labels'/f'{o.stem}_P{o.suffix}',
    codes = np.loadtxt(path/'codes.txt', dtype=str)
,num_workers=0)
learn = unet_learner(dls, resnet34)#Using resnet34 architecture to learn the model
learn.fine_tune(8)#Training the model with 8 epochs
````

````python
Now,after training the model for 8 epochs,we can observe that with every epoch training and validation
loss decreases.Now its important to understand that semantic segmentation is a pixel wise classificat-
ion and that is why it labels every pixel.It labels each pixel with the class they represent.So 
similar class pixels would be represented by same color.Also the output we get itself is a high 
resolution where each pixel represents a class.We also visualize our model predictions using
show_results.It shows 3 images and their predictions.We can see that cars are pooled in same color and
trees in different , the road crossings also , buildings and sky.It is because these are objects 
belonging to different classes.We can observe from the targets and predictions that some of the 
objects from the same class are not classified.So our model predictions are not very great here.
````

````python
#Showing results
learn.show_results(max_n=6, figsize=(7,8))
````

````python
Natural language processing is a popular field which involves interaction between computers and human
language.It involves speech processing into text,language translation,predicting answers,analyzing 
comments/posts on social media etc.Deep learning algorithms are mainly used for this purpose.So here 
in this example we are trying to build a model for sentiment analysis ie.prediction of text if it 
shows positive comments or negative.We are using IMDB reviews dataset for the prediction.
````

````python
The steps for building the model still remain the same.Fastai.text module is used for building NLP 
models so we import that. We download and decompress the data using untar_data,we already have a test
dataset as validation data and we use a batch size of 32 to divide our data into batches.We define a 
learner for the model using AWD_LSTM architecture which is particulary used for sequential data which 
requires the model to remember data trained in previous time step or instance.That is why the archite-
cture is called Long Short Term Memory models(LSTM) as it tends to have the memory of previous words 
in the data which is used for predictions later in the sequence.Finally the model is fine tuned using 
4 epochs and a learning rate of 0.01.
````

````python
#Text Prediction using fastai(Dataset-IMDB movie reviews)
from fastai.text.all import * #importing classes from fastai.text module for text processing
dls = TextDataLoaders.from_folder(untar_data(URLs.IMDB), valid='test',bs=32,num_workers=0)
#Loading the txt data and decompressing the dataset Batch size=4
learn = text_classifier_learner(dls, AWD_LSTM, drop_mult=0.5, metrics=accuracy)
#connecting the text data(dls) and AWD_LSTM model
learn.fine_tune(1, 1e-2)#training the learner with 4 epochs and learning rate
````

````python
Lets move to something which is very commonly used but not common for deep learning.All of us have
come across tabular data which consists of some numerical and some categorical variables and we 
use different Machine Learning algorithms to predict our target variable.Here we are using Adult
dataset to predict a person's income using his socioeconomic background.We follow the same process as
earlier.First we import the tabular module of fastai.Then we download and decompress the data using 
untar_data.Next we load the data in Dataloaders using the path.We also specify categorical variables(
cat_names),continuous variables(cont_names) and preprocessing process(procs).Since our model can only
handle numerical values so our preprocessing includes encoding categorical variables(categorify),
filling missing values(FillMissing) and Feature Scaling(Normalize).In Machine Learning algorithms
what requires long codes,here we can do those tasks in very few lines of code.Next we define the 
tabular_learner and we use accuracy as a metric to evaluate our model.
We are not using any pretrained model for this as there is no model available for tabular data.So we
do not fine_tune our model.Instead we use fit_one_cycle which is used for models from scratch.
````

````python
#Salary prediction from tabular database using fastai
from fastai.tabular.all import * #importing tabular module from fastai
path = untar_data(URLs.ADULT_SAMPLE)#downloading and decompressing the dataset
#Loading data into DataLoaders using the path
dls = TabularDataLoaders.from_csv(path/'adult.csv', path=path, y_names="salary",
    cat_names = ['workclass', 'education', 'marital-status', 'occupation',
                 'relationship', 'race'],#Categorical variables
    cont_names = ['age', 'fnlwgt', 'education-num'],#continuous variables
    procs = [Categorify, FillMissing, Normalize])#Preprocessing process for dataset

learn = tabular_learner(dls, metrics=accuracy)#Creating the learner using dataloaders
#No pretrained model for tabular data
````

````python
learn.fit_one_cycle(3)
#Deep learning is not very proficient for Tabular data.It's used only in case of high cardinality
#features such as product_id.Below we can observe that model is 80% accurate and takes 30 sec to run.
#We can make better predictions by training the tabular data on any Machine Learning algo such as 
#Random Forests or Naive bayes.
````

````python
We had discussed earlier about recommendation networks which is one of the applications of deep 
learning.It is mostly used by ecommerce companies to recommend products to customers.Here we are
training a model which will predict movies people may like based on the ratings given previously.Here
also we follow the same procedure.We download and decompress dataset in path.We also import fastai.
collab module which enables us to work with recommendation networks.We load the data into DataLoaders
using the path.Next we create a collab_learner for training our loaded data.We also specify y_range 
since we are predicting a continuous variable.Here also we are not using any pretrained model still we
use fine_tune method to train our model because fast.ai enables us to use fine_tune in anyways.
````

````python
#Collaborative filtering model for movie rating prediction
from fastai.collab import *
path = untar_data(URLs.ML_SAMPLE)#Downloading the path to dataset and compressing it
dls = CollabDataLoaders.from_csv(path/'ratings.csv')#Loading the data in dataloader
learn = collab_learner(dls, y_range=(0.5,5.5))#Training the model for rating prediction
learn.fine_tune(10)#Tuning the modelin 10 epochs
````

````python
#Next we use show_results to show some examples of actual ratings and predicted ratings.
learn.show_results()
````

````python
#Selection of validation and test sets:-
We know that to evaluate our model performance we need to validate and test our model on test set and 
validation set.So our data is mainly split into train set and validation test.Train set is used for 
training model and validation test is used for tuning hyperparamters mainly.Our data split should be 
such that validation set should be fully unknown to trained model so that it is more generalized.
Now even validation set is not enough since there is a chance of model getting overfit on validation 
set since we do our weight assignment on the basis of error_rate and accuracy metrics.So we reserve 
another set of data called test data.There should be enough data so as to get a good measure of model
performance as the metric specified.
Test set and validation sets are indicators of the unknown data model can see in future.Very often 
test and validation sets are chosen at random but not always this can be done.For example if there is 
a time series data you cannot randomly chose test or validation sets from the data from random dates.
In that class always earlier data is used for training and later data is used for validation.
Sometimes we need to remember special considerations while preparing our data for model building:-
For example if in training data we have a person in different positions performing different actions
with class labels such as eating,sleeping,driving,working etc..So in validation set we would want to 
have different people with different postures to get better and reproducible results.This condition
should be kept in mind while preparing datasets specially the image ones.
````
