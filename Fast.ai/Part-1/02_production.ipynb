### The Drivetrain Approach

```python
The drive train approach seems a complex term right!! But it is not.There are many models existing 
in this world which are accurate but not very useful whereas some models which are not that accurate
but used very widely.The drivetrain approach uses this concept only.It was introduced by Jeremy only
in 2012.It basically involves some sequential steps to solve a problem.Initial steps include:
1.Defining our objectives.
2.Inputs in our reach.
3.Data which can be collected
4.And finally we build the model which performs steps to achieve the objectives defined in begining.
If any of us has done udacity Machine Learning course, we should know about self driving cars built by
Google using AI.They explain the whole course through the Self driving car example.It has used 1000s 
of models and pentabytes of data to achieve this task.Predictive modelling or regression is always 
an option for the same but it has some problems.A driver driving the car is mostly unaware of the 
systemic design and the models used to build and automate car functionality.
```

```python
As Data Scientists we tend to follow a systemic design approach to achieve any defined objective.Data
is considered as the basic component of this design.We eventually use data to generate more data as 
well as a generalized,predictable outcome which a person can predict just by processing input.This is 
the core of Machine Learning.Drivetrain approach also follows the same criterion.We define an
objective for our problem we want to solve.A very simple example of which is when Google was planning 
to launch their search engine,they came up with a thought that What is the user main concern in using 
a search engine?.Obviously to get the most relavent search result for the thing they need.Next they 
thought what actions(levers) they could take to achieve this objective.They come up with a technique
to rank the search results.Now what data would they need to decide these rankings.It was only after
these steps the predictive modelling comes into picture.So we should have clear defined objectives,
the actions to achieve those objectives and the data we have and more data which we will need to build
the models.
```

```python
Let us see the example of Recommendation networks..As discussed before the main objective of 
recommendation networks is to increase the sales by recommending the customers the products they have 
not viewed or brought recently.The lever(action) here is to produce ranking of the recommendation.Lots
of new data has to be collected in order to make more sales.The company thus has to collect a lot of 
information from the customers about their browsing history,viewing history,their geography and much
more.This is done by many ecommerce companies such as amazon,flipkart etc.

So we would build 2 models for purchase probabilities,conditional on seeing or not seeing the recomme-
ndation.We have seen that practical implementation of models is way more complex than training the 
model.Right from collecting data to incorporate models into the actual system/product we are 
developing is a tedious process.Further in this notebook we would explore more about gathering data
for building models and cleaning the data..
```

## From Data to DataLoaders

```python
So next we create DataLoaders..DataLoaders is a special class which stores DataLoader object passed
and splits it into train and valid set..We can access the sets using dls.train and dls.valid.We create
DataLoaders for every type of data for which we are building the model.(Tabular,vision,text,Collabor-
ative filters) etc.Next we would also learn about Dataset and Datasets..To load our downloaded data in
DataLoaders object,we need to specify the kind of data(text,tabular,collab,vision),list of items,
get labels for the images(items) and fraction and split criteria for training and validation set.
```

```python
Now most of us would be unaware of the fact that fastai has DataBlock API system,It allows us to 
specify every process and method required to create DataLoaders object.
```

```python
We Create a DataBlock object and pass different arguments for different processes of Dataloading:-
1.blocks=(ImageBlock,CategoryBlock) specify types of inputs(independent variable) and outputs(depend- 
ent variable) for the model we are building
Here since we are building an Image Classifier so our input(independent variable) is ImageBlock and 
output(dependent variable) is CatgoryBlock because we are classifying images into different types of 
bears.
2.get_items:-Here generally we have to specify the path of our dataset.But it is underlying.Previously
we have seen that after passing path through get_image_files,we store image in a variable.So here we 
use get_image_files and it returns list of all the images in that path.
3.splitter:-Most of the times datasets have prespecified training and validation sets in different 
folders or in a csv file with a different variable called is_valid which is True for validation set 
entries and false for training.As such fastai provides us facility to randomly split the training and 
validation set randomly using RandomSplitter(valid_pct=0.2,seed=42) passing fraction of validation and
seed as arguments.Random seed so as to have a generalised result.
4.get_y=In coordinate geometry or in basic algebra also we use x,y for specific variables only though
they are just alphabets.x stands for independent variable or a set of them and y is the dependent 
variable.y=f(x).y is always depicted as a function of x.Here we specify the function to get the output
or labels.Here we downloaded our images in the specific subfolders with bear type name.Therefore we
pass parent_label which gets us the name of the folder in which the files are and used for labelling
images.
5.item_tfms=When we download images they are of different size,contrast and brightness.This can cause
problem in our model building.Since we provide images to the model in minibatches of equal sizes,We 
transform all the images into same size and therefore we pass Resize(128) and in argument the size we
want the images to be of.item_tfms refers to the transformations made to the image before it is fed to
the model.
```

```python
#create data loaders to assemble data in a format suitable for training.It also enables us to split 
#data into validation and training set,get labels and transform the images
bears = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128))
#Dataloader provides batch of items at a time to GPU
```

```python
Above command gives us the DataBlock object by passing the above arguments.Now we can create
dataloaders from this object by passing path to our images through the dataloaders.
```

```python
#Creating DataLoaders
dls = bears.dataloaders(path)
```

```python
A dataloaders mainly has a training and a validation set.It allows to feed data to the GPU in the form
of minibatches.By default the batch size is 64 but we can increase or decrease it by specifying "bs"
argument.A single batch items are all stacked up in a tensor(array like object in Pytorch).We can 
display a few items in a batch using show_batch method on a DataLoader object.
dls.train-Training set
dls.valid-Validation set
```

```python
#Displaying the few items in validation set
dls.valid.show_batch(max_n=4, nrows=1)
#WHen downloaded these images would have been of different sizes but Resize by default crops them into
#square sizes.Sometimes this can lead to loss of important information also.Like some imp features 
#such as eyes or face in the image.Fastai provides several alternatives for the transformations:-
```

```python
Some of the transformations include:Stretching or squishing,padding images with zero(black margin).
```

```python
#Stretching the images(squish)
bears = bears.new(item_tfms=Resize(128, ResizeMethod.Squish))
dls = bears.dataloaders(path)
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
#Padding the image with zeros(horizontally or vertically) that is black background
bears = bears.new(item_tfms=Resize(128, ResizeMethod.Pad, pad_mode='zeros'))
dls = bears.dataloaders(path)
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
Now the approaches we discussed above are not much useful as it can result in images looking different
from their original view and therefore can result in low accuracy.When cropping images as discussed 
before can result in losing distinctive features such as eyes or nose from the images which are imp 
for classification.Padding images with black margins results in lower resolution for the important 
part of the image.
"RandomResizedCrop" is a transformation provided by fastai and is recommended for Image Classification
.In this randomly a part of image is selected and cropped till that part.For every epoch,this part is
randomly selected and this leads to model focussing on different features of the same image and making
model more generalized so that even if any test image has different direction,color,resolution,frame..
our model can recognize the important features in the image distinctive for classification.
An untrained model cannot recognize the same object in the image if the image is rotated by 1 degree 
also.Only the direction,size of the object in the image changes.Rest remains the same!.It is nice to
train our neural network with different postures,colors,resolution of the same object in different 
images so that,it helps model in recognizing image in different conditions.It is also called Data
Augmentation in general.
In the next cell we pass RandomResizedCrop through which we pass the size for resizing and very imp
parameter min_scale which specifies the min fraction of the image to be selected each time during 
an epoch.
```

```python
#Rescaling the image using a randomized crop min_scale refers to the minimum part of
#image to be selected
bears = bears.new(item_tfms=RandomResizedCrop(128, min_scale=0.3))
dls = bears.dataloaders(path)
dls.train.show_batch(max_n=4, nrows=1, unique=True)#Displaying a batch of the dataloaders minibatch
#We can observe here 4 different items of the same object with different parts are randomly selected 
#and cropped.unique=True is used to get differeny views of the samen image.It is called data 
#augmentation
```

### Data Augmentation

```python
As discussed earlier,data augmentation is necessary for building generalized models and better valid-
ation accuracy.Data Augmentation techniques include rotating image,flipping the image,changing bright-
ness,resolution of the image so as to create verious variants of the image.For natural photo images 
,standard set of transformations work well and those are included in the "aug_transforms" function.
Once we resize the images item wise then these transformations can be applied on the whole batch and
therefore in the below cell it is applied in "batch_tfms".We also define mult=2 which says the double
the amount of default augmentations is used.These transformations are applied to whole batch on GPU
so that it does not consume more memory and it is fast.Next we load the data into dataloaders by 
passing path and we display 8 differently augmented images of the same object."unique=True" is passed
so as to get different images for every augmentation. 
```

```python
#Batch transformations applied for Image augmentation(double augmentation used compared to default)
bears = bears.new(item_tfms=Resize(128), batch_tfms=aug_transforms(mult=2))
dls = bears.dataloaders(path)
dls.train.show_batch(max_n=8, nrows=2, unique=True)
#Now since the images are in every form after applying transformations and are generalized ,model can 
#be trained to build an image classifier
```

## Training Your Model, and Using It to Clean Your Data

```python
Now since we have augmented our images,cleaned corrupt images we have also applied transformations on
the images so we can train our image classifier on it..
Since we do not have a lot of images in our data so we do a RandomResizedCrop and also apply default
aug_transforms() batch wise.Next we create dataloaders by passing path for the data.We also add 
"num_workers=0" since we are using GPU on windows.
```

```python
#Since the data is less(150 images) so randomresized crop to 224px is used to train our model
bears = bears.new(
    item_tfms=RandomResizedCrop(224, min_scale=0.5),
    batch_tfms=aug_transforms())
dls = bears.dataloaders(path,num_workers=0)
```

```python
Now we can train our learner by passing dataloaders object,the pretrained resenet18 architecture,
error_rate as metrics and we fine tune it using 4 epochs.We can see that at the end of 4th epoch , we
have an error rate of 1.2% which is very good.
```

```python
#creating learner and fine tune it at 4 epochs to train our model
learn = cnn_learner(dls, resnet18, metrics=error_rate,num_workers=0)
learn.fine_tune(4)
```

```python
Now since there is an error rate of 1.2%.This indicates that our model is wrongly classifying some
images.Let us see where it is making mistakes.All of us who have worked on machine learning we know 
about confusion matrices used for interpreting classification results.It tells us about the number of
actual and predicted classes for all categories in the vocab.Let us display confusion matrix for our
Image Classifier and see where it is making mistakes...
```

```python
#confusion matrix for classification results interpretation
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
From the confusion matrix we can see that our model is misclassifying "black" bear as "grizzly".The 
rows represent the Actual classes and columns represent Predicted classes.Though our model is working
well but it is important for us to see where our model is making mistakes.And what is the reason for 
the same..In case of images it is mainly due to images being labelled incorrectly or if it is due to
model problem.Fastai actually provides us function to plot images which have higher loss..The loss is 
high if the model is incorrect or making an incorrect classification.In next chapters,we would see how
this loss is calculated.We use "plot_top_losses" to display images with top losses.Now as it is
displayed,"Prediction/Actual/Loss/Probability"the title says every image has 4 things in the heading
including the actual and predicted labels,the loss and the probability.Here Probability is the 
confidence with which model predicts the class..
```

```python
#displaying wrongly predicted labels
interp.plot_top_losses(4, nrows=1)
```

## How to Avoid Disaster

```python
As we had discussed before also a deep learning model is a small version of a big problem system.
Developing a data product is an end to end long process from idea to final deployment.We would discuss
some of the important things which should be kept in mind while solving a problem through deep 
learning so that we can avoid chances of the failure of the project.
Understanding and debugging the behavior of a deep learning model after building it is much more
complex than that for software systems.In case of software it is easy to analyze each step and study
where we are going wrong.With deep learning models it is difficult, since it trains from the data.
While training our data we do not know the patterns in training data and how it would perform.Now 
today we built a "Bear Classifier" from the images downloaded using "Bing_Image_Search".Suppose we 
want to use that model to build a bear detection system at a National park which would alert tourists
if any bear is around by the video visuals captured from cameras around the park.However we certainly
cannot use the model we built today for this purpose because:
1.This model is trained on Image data and not on video data.
2.National park system would be required to train on night images on which this model is not trained.
3.Rapid results.
4.Identifying bears in such locations in forests.Such situational images are not there in the training
data for the model we made.
Since the model was built using the images uploaded on internet which are mostly artistic and good 
images and cannot be used as an input for building this type of bear detection system.Extensive data
collection and labelling is needed for building this system.This is called "out of domain data"
problem.And therefore we should be careful whenever we are planning to bring out any product using 
such models.Training data can be different from the data model will encounter while in production.
Another problem called "domain shift problem" also arises.When our data keeps changing with time.So 
our model also needs updation with time.This is mostly faced by the Finance and Insurance companies
when they try to do risk modelling using deep learning.Since the preferences of the customers and
type of people,their nature changes.
As such there are no solutions to this problem since neural networks are complex and we cannot analyze
every layer individually.With a careful high level approach we can limit the risks of the failure of 
the model.It basically involves human intervention along with model functioning in side.As we had 
described before,deep learning can solve such complex problems which otherwise would be very time
consuming if done by humans.But we should not completely rely on deep learning atleast not in initial
phases of production for making major decisions or taking actions.Instead a few teams could be working
on reporting where model is not able to perform or giving wrong results.Then gradually if the model
is working fine and results are as expected than we can start becoming dependable on it.But atleast 
for sometime serious human involvement and repeatative checks are necessary.
```

### Unforeseen Consequences and Feedback Loops

```python
Sometimes models can change the behavior of the environment they operate in.This happens gradually and
comes into realization very late."Predictive Policing" algorithm by US police predicts crime in areas
so that more police forces can be sent there to control crime.This results in more crimes being regi-
stered in those areas.
This is a case of feedback loops which leads to model being biased.This is happening already as acc to
reports,black people are more likely to get arrested than whites in US.And this has happened due to
"predictive policing" algorithm.This example shows that while rolling out any Machine Learning system,
a thorough analysis should be made about all the possible predictions.And a plan with actions to be 
taken in particular situations according to the model's performance should be made.
```
